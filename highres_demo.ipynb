# HighResMIP example

```python
from xmip.utils import google_cmip_col
import hvplot.xarray

col = google_cmip_col()
cat = col.search(
    activity_id='HighResMIP',
    table_id='3hr',
    grid_label='gn',
    source_id='HadGEM3-GC31-HM',
    variable_id='tas'
)
cat
```

```python
ddict = cat.to_dataset_dict()
```

```python
ds = ddict['HighResMIP.MOHC.HadGEM3-GC31-HM.highresSST-present.3hr.gn']
ds
```

```python
da = ds.tas.squeeze()
da
```

```python
da.hvplot(groupby='time', width=600, widget_type='scrubber', widget_location='bottom', rasterize=True)
```
